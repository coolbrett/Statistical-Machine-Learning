```python
import pandas as pd
```

```python
housing_df = pd.read_csv("https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv")
```

```python
housing_df.head()
```

```python
housing_df.drop(columns=['longitude'])
housing_df.head()
```

```python
housing_df = housing_df.dropna(axis=0)
housing_df = housing_df.drop(columns=['ocean_proximity'])
housing_df.info()
```

```python
housing_df.to_csv()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import f_regression
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
five_feature_selector = SelectKBest(score_func=mutual_info_regression, k=5)
lm5 = LinearRegression()
lm5_pipeline = make_pipeline(five_feature_selector, lm5)

three_feature_selector = SelectKBest(score_func=f_regression, k=3)
lm3 = LinearRegression()
lm3_pipeline = make_pipeline(three_feature_selector, lm3)
```

```python
y_var = housing_df["median_house_value"]
x_var = housing_df.drop(columns=["median_house_value"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_var, y_var, train_size=0.7, random_state=23)
```

```python
lm5_pipeline.fit(x_train, y_train)
```

```python
lm3_pipeline.fit(x_train, y_train)
```

```python
house_12 = x_train.loc[12]
```

```python
lm3_pipeline.predict([house_12])
```

```python
lm3_residual_12 = y_train[12] - lm3_pipeline.predict([house_12])
lm5_residual_12 = y_train[12] - lm5_pipeline.predict([house_12])
```

```python
lm3_residual_12
```

```python
lm5_residual_12
```

### so the 5 feature residuals are much higher

```python
lm3_pipeline.score(x_test, y_test)
```

```python
lm5_pipeline.score(x_test, y_test)
```

### The 5 feature scores are considerably higher and closer to 1 with the testing data

```python
lm3_pipeline.score(x_train, y_train)
```

```python
lm5_pipeline.score(x_train, y_train)
```

### None of these models are that great, but if I had to recommend, I would say the 5 feature model.
